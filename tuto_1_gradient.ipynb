# Algorithmes de descente de gradients

"L'algorithme du gradient" est un algorithme d'optimisation destiné à minimiser une fonction réelle différentiable. C'est un algorithme  itératif qui procède donc par améliorations successives. 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## La descente de Gradient: un exemple simple

Nous considérons le problème de trouver un minimum d’une fonction $ f $, donc de résoudre
$$\umin{x \in \RR^d} f(x)$$ ou $f : \RR^d \rightarrow \RR$ est une fonction derivable.

Notez que le minimum n'est pas nécessairement unique. $ f $ pourrait eventuellementprésenter des minima locaux, auquel cas l'algorithmes  pourrait ne pas trouver un minimiseur global du problème. Pour commencer, nous limitons notre attention à une simple fonction convexe, pour eviter ce probleme. La descente de gradient calcule:
$$ x ^ {(k + 1)} = x ^ {(k)} - \tau_k \nabla f (x ^ {(k)}), $$
où $\tau_k> 0 $ est une taille de pas et $ \nabla f (x) \in \RR ^ d $ est le gradient de $ f $ au point $ x $ et de $ x ^ {(0)} \in \RR ^ d $ est un point initial.

Nous considérons un problème simple, correspondant à la minimisation d'une forme quadratique bi-dimensional 
$$ f(x) = \frac{1}{2} \pa{ x_1^2 + \eta x_2^2 }$$.

```python
eta = 4
f = lambda x : ( x[0]**2 + eta*x[1]**2 ) / 2
```

Essayons d'abord de regarder a quoi ressemble cette fonction:

```python
tx = np.linspace(-.3,1,101)
ty = np.linspace(-.6,.6,101)
[v,u] = np.meshgrid(ty,tx)
F = ( u ** 2 + eta * v ** 2 ) / 2
plt.contourf(tx,ty,F.transpose(),10);
```

Nous allont maintenant calculer le gradient de cette fonction. Il s'agit bien evidement d'un vecteur a deux dimensions:
$$\nabla f(x) = x_1 \vec e_1 + \eta x_2 \vec e_2, $$.

Nous ecrivons la fonction associee

```python
Gradf = lambda x : np.array([x[0],eta*x[1]])
```

puis nous allons coder la descente de gradient:

## L'algorithme de Newton

La méthode de Newton tente de trouver la solution de $f'=0$ en construisant une séquence de facon iterative. En supposant que $f'(x)=g(x)$, on ecrit 
$$
g(x^{t+1}) = g(x^t+\delta x) \approx g(x^t) + \delta x g'(x^t)
$$
et donc la condition $g(x)=0$ conduit a 
$$ \delta x = -\frac{g(x^{t})}{g'(x^t)} = -\frac{f'(x^{t})}{f''(x^t)}$$

L'algorithme de Newton s'ecrit donc, a une dimension:
$$ x ^ {(k + 1)} = x ^ {(k)} - \frac{f' (x ^ {(k)})}{f'' (x ^ {(k)})}, $$
et en dimension superieure
$$ \mathbf {x} _{n+1}=\mathbf {x} _{n}-[\mathbf {H} f(\mathbf {x} _{n})]^{-1}\nabla f(\mathbf {x} _{n}),\ n\geq 0. $$

Son inconvenient est qu'il necessite le calcul est l'inversion de la Hessienne. Il est pourant bien plus effiace que la descente de Gradient comme nous allons le voir:

```python
InvHessian = np.array([[1,0],[0,1/eta]])

niter = 3
x = x0
Values=[]

X = np.zeros((2,niter))
for i in np.arange(0,niter):
    Values.append(f(x))
    X_history[:,i] = x.flatten()
    # optimal step size
    x = x - InvHessian@Gradf(x)
plt.clf
plt.contourf(tx,ty,F.transpose(),10);
plt.plot(X_history[0,:], X_history[1,:], 'k.-');
plt.axis('equal')
```

```python
plt.clf
h = plt.plot((Values))
plt.axis('tight')
plt.title('Valeur de la function')
```

Ici, nous avons trouve le bon minimum est une seule iteration! Mais il faut dire que la fonction etait particulierment simple

Nous considérons un problème plus difficile:
$$ f(x) = \pa{ x_1^4 -x_1^2 + \eta x_2^2 - 2 * x_1 x_2}$$.

```python
eta = 4
f = lambda x : ( x[0]**4 -x[0]**2+ eta*x[1]**2 -2*x[0]*x[1])
```

```python
tx = np.linspace(-2,2,101)
ty = np.linspace(-2,2,101)
[v,u] = np.meshgrid(ty,tx)
F = ( f([v,u])) 
plt.contourf(tx,ty,F.transpose(),10);
```

```python
Gradf = lambda x : np.array([4*x[0]**3-2*x[0]-2*x[1],2*eta*x[1]-2*x[0]])
```

```python
niter = 2000

plt.clf
plt.contourf(tx,ty,F.transpose(),10);
Values=[]


tau_list = np.array([0.2]) 

for itau in np.arange(0,tau_list.size):
    tau = tau_list[itau]
    x = x0
    X_history = np.zeros((2,niter))
    for i in np.arange(0,niter):
        Values.append(f(x))
        X_history[:,i] = x.flatten()
        x = x - tau*Gradf(x)
    plt.plot(X_history[0,:], X_history[1,:], '.-');
```
